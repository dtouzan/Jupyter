## Introduction
Les observations ont été effectuées avec le télescope Intes M603. La caméra CCD utilisée est la Atik Titan la configuration de celle-ci est de 659 x 494 pixels de 7,4 µm avec un refroidissement non régulé. La résolution avec le télescope Intes M603 d’une focale de 1 mètre est de 1,6" arc sec / pixel, Un réducteur de focale Celestron est utilisé pour pouvoir descendre la focale à 1 mètre (F/D 6,7). La monture est une Sky-Watcher AZ-EQ5. Le suivi automatique étant assuré par un télescope LC90 et une caméra PL-1M. Les prises de vue sont réalisées essentiellement en début de nuit entre vingt-deux heures et minuit.
 
Le logiciel CCDCiel est utilisé pour la prise de vue, pour le suivi automatique PHD2 est nécessaire (binning deux avec la PL-1M) pour un bon suivi de la monture. Les temps de pose cumulés d’une heure cumulé sont faciles à réaliser avec la monture. Les images sont  réalisées en utilisant la fonction « médian » sous le logiciel Iris[[1]](#ref1). L’orientation des images est calibrée le nord en haut et l’est à gauche. Les prévisualisations des images ont été élaborées avec le logiciel Aladin pour le repérage des étoiles variables et la calibration astrométrique.

## Hauteurs de l'étoile variable à l'observatoire UT1.
La hauteur de v1492 Cyg vers 20h UT est proche du zénith, la constellation du cygne est bien accéssible pour le site cependant certaines soirées sont délicates car le temps ne se prête guère à l'observation et la Lune ne permet pas de bonnes prises de vue quand celle ci est illuminée au dessus de 25%. Le graphique _(Fig. 2.)_ montre l'élévation de l'astre ainsi que de la Lune pour la période des prises de vue.

```python
# Dates des observations
dates_obs = ['2019-08-29T20:00:00','2019-08-30T20:00:00','2019-09-05T20:00:00','2019-09-08T20:00:00',
             '2019-09-10T20:00:00','2019-09-12T20:00:00','2019-09-14T20:00:00','2019-09-15T20:00:00',
             '2019-09-17T20:00:00','2019-09-18T20:00:00','2019-09-19T20:00:00','2019-09-20T20:00:00',
             '2019-09-22T20:00:00','2019-09-27T20:00:00']

# % illumination de la Lune
moon_phase = [0,0,25,50,50,75,100,100,75,75,50,50,25,0]

Elevationsplot('v1492cyg', dates_obs, moon_phase, 25)
```

_Fig. 2._ Hauteur de v1492 Cyg et illumination et hauteur de la Lune durant les quatorze soirées.
