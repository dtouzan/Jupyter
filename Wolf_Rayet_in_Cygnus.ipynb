                    <b>Traitements </b> <br><br>
                    La nécéssité de visualiser le rendu des spectres avant la prise de vue de ceux ci est très utile pour ce genre de spectrographe car des étoiles dans le chemin du spectre provoque de bon artefacts. Le logiciel utilisé est Aladin Sky Atlas V11 qui permet de faire la rotation des champs selectionnés, pour cela le champ du spectre a été créé et importé avec cet outil. Le traitement des images est une médiane de minimum douze images pré-traitées (noir, bias et flat) puis d'un découpage du spectre incluant l'étoile d'ordre 0. Le retournement horizontal de l'image pour le traitement du spectre sur Isis est nécessaire. 
                    <br><br>
                    Pour la réalisation de la courbe de chaque spectre le traitement c'est fait par étalonage par la dispersion, un découpage de 420 nm à 670 nm, la fabrication de la courbe du continuum et division le la courbe étalonée par celle ci. <br><br>

```python
import warnings
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
from matplotlib import colors
import numpy as np
from astropy import wcs
from astropy.io import fits
from astropy.utils.exceptions import AstropyWarning
from astropy.visualization import quantity_support
import ciboulette.base.ciboulette as Cbl
from ciboulette.utils.archive import Archive
from ciboulette.spectrum import sa200
import ciboulette.base.constant as C

quantity_support()
warnings.simplefilter('ignore', category=AstropyWarning)

cbl = Cbl.Ciboulette()

cbl.canon200_2_8
cbl.asi178
```

```python
# Serie de balmer
Hg = 4340.46               # BLUE COLOR
Hb = 4861.32               # CYAN COLOR
Ha = 6562.10               # RED COLOR
# Raie Wolf-Rayet
HII_1 = 4686               # BLUE COLOR
HII_2 = 5411               # GREEN COLOR
CIII = 5696                # YELLOW COLOR
CIV = 5800                 # ORANGE COLOR

wr_stars = ('HD190918', 'HD191765', 'HD192103', 'HD192163', 'HD192641', 'HD193077', 'HD193576', 'HD193793')

dataset = ('dataset/WR2022/Pc_hd190918-wr133_20220826_852.fits',
           'dataset/WR2022/Pc_hd191765-wr134_20220819_862.fits',
           'dataset/WR2022/Pc_hd192103-wr135_20220819_920.fits',
           'dataset/WR2022/Pc_hd192163-wr136_20220819_862.fits',
           'dataset/WR2022/Pc_hd192641-wr137_20220826_852.fits',
           'dataset/WR2022/Pc_hd193077-wr138_20220819_892.fits',
           'dataset/WR2022/Pc_hd193576-wr139_20220826_903.fits',
           'dataset/WR2022/Pc_hd193793-wr140_20220827_848.fits')

datafits = ('dataset/WR2022/sHD190918.fits',
           'dataset/WR2022/sHD191765.fits',
           'dataset/WR2022/sHD192103.fits',
           'dataset/WR2022/sHD192163.fits',
           'dataset/WR2022/sHD192641.fits',
           'dataset/WR2022/sHD193077.fits',
           'dataset/WR2022/sHD193576.fits',
           'dataset/WR2022/sHD193793.fits')

fieldfits = ('dataset/WR2022/HD190918-20220826-2113-7x180s-f400.fits',
           'dataset/WR2022/HD191765-20220826-2113-7x180s-f400.fits',
           'dataset/WR2022/HD192103-20220826-2113-7x180s-f400.fits',
           'dataset/WR2022/HD192163-20220819-2126-7x180s-f400.fits',
           'dataset/WR2022/HD192641-20220826-2038-8x180s-f400.fits',
           'dataset/WR2022/HD193077-20220819-2126-7x180s-f400.fits',
           'dataset/WR2022/HD193576-20220826-2142-8x180s-f400.fits',
           'dataset/WR2022/HD193793-20220827-2016-12x130s-f400.fits')

table_sa200 = []
for file in dataset:
    table_sa200.append(sa200.SA200(file))
    
```

```python
figure = plt.figure(figsize=(25,42))
plt.gcf().subplots_adjust(left = 0.1, bottom = 0.1, right = 0.9, top = 0.9, wspace = 0.3, hspace = 0.3)

index = 1
star = 0
for S in table_sa200:
    cbl.positionsbyname = wr_stars[star]
    cbl._date = S.date
    P = cbl.projections
    P.title = wr_stars[star]
    M = cbl.starsmap

    axes1 = figure.add_subplot(8,5,index,projection = 'aitoff') # Renvoie un objet AxesSubplot, sous classe de Axes
    P.plot(axes1)

    axes2 = figure.add_subplot(8,5,index+1,projection = M.WCS) # renvoie un objet Axes
    M.plot(axes2)
    
    axes3 = figure.add_subplot(8,5,index+2) # renvoie un objet Axes
    image_data = fits.getdata(datafits[star], ext=0)
    mean = np.mean(image_data)
    axes3.imshow(image_data, cmap='Greys', vmin = mean - 3000, vmax = mean + 8000, origin = 'lower')
    axes3.set_title('Spectrum display\n'+ S.date, fontsize = 10)
    axes3.axes.yaxis.set_visible(False)

    
    axes4 = figure.add_subplot(8,5,index+3) # Renvoie un objet AxesSubplot, sous classe de Axes
    axes4.set_title(S.title, fontsize = 8)
    axes4.step(S.s1d.spectral_axis, S.s1d.flux)
    axes4.set_ylabel('Intensity'.format(S.s1d.flux.unit))
    axes4.vlines(Hg, 1, 6, linestyles = 'dashed', colors = 'blue', alpha = 0.8, label = r'H$\gamma$')
    axes4.vlines(HII_1, 1, 6, linestyles = 'dashed', colors = 'blue', alpha = 0.5, label = r'HII')
    axes4.vlines(Hb, 1, 6, linestyles = 'dashed', colors = 'cyan', alpha = 0.5, label = r'H$\beta$')
    axes4.vlines(HII_2, 1, 6, linestyles = 'dashed', colors = 'green', alpha = 0.5, label = r'HII')
    axes4.vlines(CIII, 1, 6, linestyles = 'dashed', colors = 'yellow', alpha = 0.8, label = r'CIII')
    axes4.vlines(CIV, 1, 6, linestyles = 'dashed', colors = 'orange', alpha = 0.8, label = r'CIV')
    axes4.vlines(Ha, 1, 6, linestyles = 'dashed', colors = 'red', alpha = 0.5, label = r'H$\alpha$')
    axes4.legend()
    
    axes5 = figure.add_subplot(8,5,index+4) # renvoie un objet Axes
    image_data = fits.getdata(fieldfits[star], ext=0)
    mean = np.mean(image_data)
    
    if wr_stars[star] == 'HD192163' :
        axes5.imshow(image_data, cmap='Greys', vmin = mean - 150, vmax = mean + 100, origin = 'lower')
    else :
        axes5.imshow(image_data, cmap='Greys', vmin = mean - 200, vmax = mean + 500, origin = 'lower')
   
    axes5.axes.xaxis.set_visible(False)
    axes5.axes.yaxis.set_visible(False)

   
    star += 1
    index += 5

plt.show()
```
