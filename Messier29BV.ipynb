## Introduction
Les observations ont été effectuées avec le télescope Intes M603. La caméra CCD utilisée est la Atik Titan la configuration de celle-ci est de 659 x 494 pixels de 7,4 µm avec un refroidissement non régulé. La résolution avec le télescope Intes M603 d’une focale de 1 mètre est de 1,6" arcsec / pixel, Un réducteur de focale Celestron est utilisé pour pouvoir descendre la focale à 1 mètre (F/D 6,7). La monture est une Sky-Watcher AZ-EQ5. Le suivi automatique étant assuré par un télescope LC90 et une caméra PL-1M. Les prises de vue sont réalisées essentiellement en début de nuit entre vingt-deux heures et minuit.
 
Le logiciel CCDCiel est utilisé pour la prise de vue, pour le suivi automatique PHD2 est nécessaire (binning deux avec la PL-1M) pour un bon suivi de la monture. Les temps de pose cumulés de dix minutes pour le filtre CLS sont faciles à réaliser avec la monture. Les images sont réalisées en utilisant la fonction « médian » sous le logiciel Iris[[3]](#ref3). L’orientation des images est calibrée le nord en haut et l’est à gauche. Les prévisualisations des images ont été élaborées avec le logiciel Aladin pour le repérage de l'amas M 29 et la calibration astrométrique.

```python
from astrolebaladin import *
```

```python
# Variables utilisées
object_name = 'M 29'
field = 0.20
file_BV = 'm29-ucac4.csv'
file_fits = 'm29-20190828.fits'
# Filtre type spectral O...M
Sptype = ''
# Indice B-V à mettre en évidence
indice_BV = 1.5

# Tableaux utilisés
data_HR = TableHR()
data_table = RegionBV_UT1(file_BV,data_HR)
data_field = RegionBV_UCAC4(object_name,field)
SpotHRClasses = ResultHRClasses(data_HR,data_table)
```
