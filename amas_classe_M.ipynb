## Introduction
Les observations ont été effectuées avec le télescope Intes M603. La caméra CCD utilisée est la Atik Titan la configuration de celle-ci est de 659 x 494 pixels de 7,4 µm avec un refroidissement non régulé. La résolution avec le télescope Intes M603 d’une focale de 1 mètre est de 1,6" arcsec / pixel, Un réducteur de focale Celestron est utilisé pour pouvoir descendre la focale à 1 mètre (F/D 6,7). La monture est une Sky-Watcher AZ-EQ5. Le suivi automatique étant assuré par un télescope LC90 et une caméra PL-1M. Les prises de vue sont réalisées essentiellement en début de nuit entre vingt heures et minuit.
 
Le logiciel CCDCiel est utilisé pour la prise de vue, pour le suivi automatique PHD2 est nécessaire (binning deux avec la PL-1M) pour un bon suivi de la monture. Les temps de pose cumulés d'une vingtaine de minutes pour le filtre CLS et d'une vingtaine de seconde pour les filtres B (bleu) et G (vert) sont faciles à réaliser avec la monture. Les images sont réalisées en utilisant la fonction « médian » sous le logiciel Iris[[2]](#ref2). L’orientation des images est calibrée le nord en haut et l’est à gauche. Les prévisualisations des images ont été élaborées avec le logiciel Aladin pour le repérage des amas et la calibration astrométrique.

## Observation.

Les images de la soirées ont d'abord été calibrées avec le logiciel **Iris** puis additionées, pour celles-ci une addition de 15 ou 20 images de 60 secondes de la soirée du 15 février 2020. Une calibration astrométrique est appliquée pour indiquer les coordonées **WCS** avec le logiciel **Aladin Sky Atlas**. Sur les images en négatif les étoiles aux alentours de la classe M ont été repérées. Les cartes de champs de ces amas sont réalisées avec **Astroquery** et deux catalogues disponibles avec **VizieR**, les catalogues **UCAC4** et **APASS**. Les indications de magnitude ont été reportées.

<B>NGC 1807.</B> Amas bien éparpillé avec une ligne définie nord sud, une seule étoile approchant le type spectral M dans ce champs UCAC4 533-011989. Deux galaxies repérées LEDA 16865 et 2E 1228 dans cet amas, 2E 1228 est de type Seyfert 1 et de magnitude 15.6. Ce n'est pas un amas physique d'après l'article _"uvby-Hβ CCD photometry of NGC 1817 and NGC 1807"_[[4]](#ref4)

## Report des mesures.
Le catalogue UCAC4 a été choisi. Utilisant un filtre G au lieu de V (visuel) les valeurs B-V diffères mais on retrouve bien les étoiles approchant ou étant de classe M bien séparées. Pour les amas NGC 1662 et NGC 2129 le type spectral est plus centré sur les classes F et G, pour NGC 1807 la classe spectrale K est prédominante. Un médian sur 15 images de 20 secondes de temps de pose unitaire est réalisé aussi bien avec le filtres G que le filtre B pour la prise des mesures. La magnitude pour le filtre G est calculée (Formule de l'AAVSO) à partir des resultats du filtre B et de la valeur Bmag
du catalogue UCAC4

$$mag_{1}-mag_{2}=-2,5.\log_{10} (\frac{flux_{1}}{flux_{2}})$$

Les valeurs utilisées pour les graphes ci dessous sont celles du catalogue UCAC4 et les mesures des étoiles selectionnées dans les tables 2,3 et 4.

```python
data_HR = TableHR()
data_NGC = Table([['NGC 1807','NGC 1662','NGC 2129'],
                  ['ngc1807-ucac4.csv','ngc1662-ucac4.csv','ngc2129-ucac4.csv'], 
                  ['ngc1807-20200115.fits','ngc1662-20200115.fits','ngc2129-20200115.fits']], names=['NGC_name', 'CSV', 'FITS'])

for NGC in data_NGC:
    data_table = RegionBV_UT1(NGC['CSV'],data_HR)
    data_field = RegionBV_UCAC4(NGC['NGC_name'],field)
    SpotHRClasses = ResultHRClasses(data_HR,data_table)
    PlotBV(NGC['NGC_name'],data_table,data_field,SpotHRClasses,data_HR,NGC['FITS'],indice_BV)
```
