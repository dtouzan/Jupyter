## Introduction
Les observations ont été effectuées avec le télescope Intes M603. La caméra CCD utilisée est la Atik Titan la configuration de celle-ci est de 659 x 494 pixels de 7,4 µm avec un refroidissement non régulé. La résolution avec le télescope Intes M603 d’une focale de 1 mètre est de 1,6" arc sec / pixel, Un réducteur de focale Celestron est utilisé pour pouvoir descendre la focale à 1 mètre (F/D 6,7). La monture est une Sky-Watcher AZ-EQ5. Le suivi automatique étant assuré par un télescope LC90 et une caméra PL-1M. Les prises de vue sont réalisées essentiellement en début de nuit entre vingt-deux heures et minuit.
 
Le logiciel CCDCiel est utilisé pour la prise de vue, pour le suivi automatique PHD2 est nécessaire (binning deux avec la PL-1M) pour un bon suivi de la monture. Les temps de pose cumulés d’une heure cumulé sont faciles à réaliser avec la monture. Les images sont  réalisées en utilisant la fonction « médian » sous le logiciel Iris[[1]](#ref1). L’orientation des images est calibrée le nord en haut et l’est à gauche. Les prévisualisations des images ont été élaborées avec le logiciel Aladin pour le repérage des étoiles variables et la calibration astrométrique.

## Hauteurs de l'étoile variable à l'observatoire UT1.
La hauteur de v2492 Cyg vers 0h UT est proche du zénith, la constellation du cygne est bien accéssible pour le site avec une faible turbulence.

## Repérage de l'étoile variable dans la nébuleuse IC 5070 (Nébuleuse du Pélican).
Pour indiquer le repérage les images des soirées ont d'abord été calibrées avec le logiciel **Iris** puis additionées (addition de 41 images de 180 sec) puis avec le logiciel **Aladin Sky Atlas** une calibration astrométrique est appliquée pour indiquer les coordonées **WCS**. La très faible étoile est de magnitude inférieur à 18.50 et est difficilement observable pour l'instrument. La pose cumulée de 2 heures permet toutefois de faire apparaitre une faible luminosité qui est révélée par la courbe ADU (*Analog-to-digital unit*), celle ci est de l'ordre d'une quinzaine d'ADU supérieur au fond du ciel.

```python
# Affichage des images et valeurs ADU
# Spot(Nom CDS, image fits, seuil minimum, seuil maximum, titre, image jpeg, titre)
Spot('v2492cyg', 'v2492cyg-20190901.fits', 200, 300, '1', 'v2492cyg.jpg', 'Zoom x2, palette de couleur "rainbow inversé"')
# ADUplot(fichier ADU, titre)
ADUplot('v2492cyg_adu.csv', 'Coupe A B')
```

_Fig. 3._ Repérage de v2492 Cyg dans une partie de la nébuleuse IC 5070 le 01/09/2019 à 20h 10mn . Courbe ADU de la coupe AB.
